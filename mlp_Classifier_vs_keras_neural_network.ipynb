MLP calssifier with sklearn andthen compare it with keras neural network


```python
from tensorflow.keras.datasets import mnist
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler # Changed minmax_scaler to MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# split the data
(train_X, train_y), (test_X, test_y) = mnist.load_data()

# faltten the data
train_X = train_X.reshape(-1, 784)
test_X = test_X.reshape(-1, 784)

# normalize the dataset
scaler = MinMaxScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```

```python
# now data is ready
#load th ecalssifier
model = MLPClassifier()
# Step 5: Define and Train the Model
model = MLPClassifier(
    hidden_layer_sizes=(128, 64),  # Two hidden layers: first with 128 units, second with 64 units
    activation='relu',             # Activation function for hidden layers
    solver='adam',                 # Optimizer (Adam works well for most cases)
    batch_size=128,                # Batch size for training
    max_iter=20,                   # Number of epochs to train the model
    verbose=True,                  # Print progress messages
    random_state=42                # Set seed for reproducibility
)

model.fit(train_X, train_y)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
# Step 6: Evaluate the Model on the Test Set
y_pred = model.predict(test_X)

# Calculate Test Accuracy
test_accuracy = accuracy_score(test_y, y_pred)
print(f"Test Accuracy: {test_accuracy:.2f}")

# Generate Classification Report
classification_rep = classification_report(test_y, y_pred)
print("\nClassification Report:")
print(classification_rep)

# Step 7: Create and Plot the Confusion Matrix
conf_matrix = confusion_matrix(test_y, y_pred)

# Plot Confusion Matrix in a Fancy Way with Color for Correct and Incorrect Predictions
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='RdYlGn', linewidths=0.5, linecolor='black')
plt.title("Confusion Matrix with Highlighted Correct and Incorrect Predictions")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

now neural network with KERAS

```python
# Step 1: Load Required Libraries
import numpy as np  # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for plotting graphs
import seaborn as sns  # Seaborn for creating fancy heatmaps
from tensorflow.keras.datasets import mnist  # MNIST dataset from Keras
from tensorflow.keras.models import Sequential  # Sequential model from Keras
from tensorflow.keras.layers import Flatten, Dense, Dropout  # Layers for the neural network
from tensorflow.keras.utils import to_categorical  # Utility to convert labels to one-hot encoding
from sklearn.metrics import confusion_matrix  # Function to create a confusion matrix

# Step 2: Load the MNIST Dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()  # Load training and test data from MNIST dataset

# Step 3: Preprocess the Data
# Normalize the Data to the Range [0, 1]
x_train = x_train / 255.0  # Normalize training data
x_test = x_test / 255.0  # Normalize test data

# One-Hot Encode the Labels
y_train = to_categorical(y_train, 10)  # Convert training labels to one-hot encoding
y_test = to_categorical(y_test, 10)  # Convert test labels to one-hot encoding

# Step 4: Define the Neural Network Model
model = Sequential()  # Initialize a Sequential model

# Input Layer: Flatten the 28x28 Images
model.add(Flatten(input_shape=(28, 28)))  # Flatten the input images from 28x28 to a 1D vector of 784 elements

# Hidden Layers
model.add(Dense(128, activation='relu'))  # First hidden layer with 128 neurons and ReLU activation
model.add(Dropout(0.3))  # Dropout layer to reduce overfitting by randomly setting 30% of inputs to zero
model.add(Dense(64, activation='relu'))  # Second hidden layer with 64 neurons and ReLU activation
model.add(Dropout(0.3))  # Dropout layer to reduce overfitting by randomly setting 30% of inputs to zero

# Output Layer
model.add(Dense(10, activation='softmax'))  # Output layer with 10 neurons (one for each class) and softmax activation for classification

# Step 5: Compile the Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])  # Compile the model with Adam optimizer and categorical cross-entropy loss

# Step 6: Train the Model
history = model.fit(x_train, y_train, batch_size=128, epochs=10, validation_split=0.2)  # Train the model with a batch size of 128, for 10 epochs, using 20% of training data for validation

# Step 7: Evaluate the Model on the Test Set
test_loss, test_accuracy = model.evaluate(x_test, y_test)  # Evaluate the model on the test data
print(f"Test Accuracy: {test_accuracy:.2f}")  # Print the test accuracy

# Step 8: Make Predictions and Plot the Confusion Matrix
y_pred = model.predict(x_test)  # Predict the classes for the test data
y_pred_classes = np.argmax(y_pred, axis=1)  # Get the predicted class for each test sample
y_true = np.argmax(y_test, axis=1)  # Get the true class for each test sample

# Create Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred_classes)  # Create the confusion matrix

# Plot Confusion Matrix
plt.figure(figsize=(10, 8))  # Set the figure size for the plot
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='RdYlGn', linewidths=0.5, linecolor='black')  # Plot the confusion matrix with annotations
plt.title("Confusion Matrix with Highlighted Correct and Incorrect Predictions")  # Set the title for the plot
plt.xlabel("Predicted Label")  # Set the label for the x-axis
plt.ylabel("True Label")  # Set the label for the y-axis
plt.show()  # Display the plot

# Step 9: Visualize Training and Validation Accuracy/Loss
# Plot Training and Validation Loss and Accuracy
plt.figure(figsize=(12, 5))  # Set the figure size for the plots

# Plot Loss
plt.subplot(1, 2, 1)  # Create the first subplot for loss
plt.plot(history.history['loss'], label='Training Loss')  # Plot the training loss
plt.plot(history.history['val_loss'], label='Validation Loss')  # Plot the validation loss
plt.xlabel('Epochs')  # Set the label for the x-axis
plt.ylabel('Loss')  # Set the label for the y-axis
plt.title('Training and Validation Loss')  # Set the title for the loss plot
plt.legend()  # Add a legend to the plot

# Plot Accuracy
plt.subplot(1, 2, 2)  # Create the second subplot for accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')  # Plot the training accuracy
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')  # Plot the validation accuracy
plt.xlabel('Epochs')  # Set the label for the x-axis
plt.ylabel('Accuracy')  # Set the label for the y-axis
plt.title('Training and Validation Accuracy')  # Set the title for the accuracy plot
plt.legend()  # Add a legend to the plot

plt.show()  # Display the plots
```
